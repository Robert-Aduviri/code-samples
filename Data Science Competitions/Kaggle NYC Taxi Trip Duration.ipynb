```python
def run_model(clf, X, y):
    start = int(time.time() * 1000)
    scores = cross_val_score(clf, X, y, cv=5, scoring='neg_mean_squared_error')
    end = int(time.time() * 1000)
    print("%s \t\t MSE: %s (+/- %s) Time: %s ms" % \
          (str(clf.__class__).split('.')[-1].replace('>','').replace("'",''), \
          locale.format("%0.2f", scores.mean(), grouping=True),
          locale.format("%0.2f", scores.std() * 2, grouping=True),
          locale.format("%d", float(end - start), grouping=True)))
          #end - start))
```

```python
def run_models(X, y):
    run_model(LinearRegression(), X, y)
    run_model(Ridge(), X, y)
#     run_model(Lasso(), X, y)
#     run_model(ElasticNet(), X, y)
#     run_model(SVR(verbose=True), X, y)
    run_model(SGDRegressor(max_iter=10), X, y)
#     run_model(KernelRidge(), X, y)
#     run_model(KNeighborsRegressor(), X, y)
    run_model(XGBRegressor(silent=False), X, y)
#     run_model(GaussianProcessRegressor(), X, y)
    run_model(RandomForestRegressor(verbose=1), X, y)
#     run_model(GradientBoostingRegressor(verbose=1), X, y)    
#     run_model(MLPRegressor(verbose=True), X, y)
```

```python
from datetime import datetime
to_weekday = ['Lunes', 'Martes', 'Miercoles', 'Jueves', 'Viernes', 'Sabado', 'Domingo']
def get_weekday(fecha):
    datetime_obj = datetime.strptime(fecha.split()[0], "%Y-%m-%d")
#     return to_weekday[datetime_obj.weekday()]
    return datetime_obj.weekday()
```

```python
from math import radians, cos, sin, asin, sqrt
def haversine(lon1, lat1, lon2, lat2):
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    km = 6367 * c
    return km
```

```python
def preprocess(df):
    df_pickup_datetime = df['pickup_datetime'].apply(lambda x: x.replace('-',' ').replace(':',' ').split())
    df['pickup_year'] = df_pickup_datetime.apply(lambda x: int(x[0]))
    df['pickup_month'] = df_pickup_datetime.apply(lambda x: int(x[1]))
    df['pickup_day'] = df_pickup_datetime.apply(lambda x: int(x[2]))
    df['pickup_hour'] = df_pickup_datetime.apply(lambda x: int(x[3]))
    df['pickup_minute'] = df_pickup_datetime.apply(lambda x: int(x[4]))
    df['store_and_fwd_flag'] = df['store_and_fwd_flag'].apply(lambda x: 0 if x=="N" else 1)
    df['pickup_weekday'] = df['pickup_datetime'].apply(get_weekday)
    df['distance'] = df.apply(lambda row: haversine(row['pickup_longitude'], row['pickup_latitude'],
                                                    row['dropoff_longitude'], row['dropoff_latitude']), axis=1)
    df = df.drop(['id', 'pickup_datetime'], axis=1)
    return df
```

```python
df_train = pd.read_csv('train.csv')
df_train = preprocess(df_train)
```

```python
df_train.head()
```

```python
from sklearn.preprocessing import StandardScaler
X, y = df_train.drop(['trip_duration', 'dropoff_datetime'], axis=1), df_train['trip_duration']

normalizer = StandardScaler()
X = normalizer.fit_transform(X)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
reg = LinearRegression()
reg.fit(X, y)
y_pred = reg.predict(X)
mean_squared_error(y, y_pred)
```
